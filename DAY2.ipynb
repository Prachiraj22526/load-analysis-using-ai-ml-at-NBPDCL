```python
candidates = [c for c in df.columns if any(k in c.lower() for k in ('date','time','timestamp','ts'))]
for c in candidates:
    df[c] = pd.to_datetime(df[c], errors='coerce', infer_datetime_format=True)
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
outlier_counts = {}
for c in num_cols:
    col = df[c].dropna()
    q1,q3 = col.quantile(0.25), col.quantile(0.75)
    iqr = q3 - q1
    lower, upper = q1 - 1.5*iqr, q3 + 1.5*iqr
    outlier_counts[c] = int(((df[c] < lower) | (df[c] > upper)).sum())
display(pd.DataFrame(list(outlier_counts.items()), columns=['column','outlier_count']).sort_values('outlier_count', ascending=False))
```

```python
cleaned = df.drop_duplicates()
if candidates:
    main_ts = candidates[0]
    cleaned = cleaned[~cleaned[main_ts].isna()]
```

```python
num_cols = cleaned.select_dtypes(include=[np.number]).columns.tolist()
cleaned[num_cols] = cleaned[num_cols].apply(lambda x: x.fillna(x.median()))
```

```python
obj_cols = cleaned.select_dtypes(include=['object']).columns.tolist()
for c in obj_cols:
    if cleaned[c].isna().any():
        m = cleaned[c].mode(dropna=True)
        cleaned[c] = cleaned[c].fillna(m.iloc[0] if not m.empty else "missing")
```

```python
def remove_iqr_outliers(df_in, cols):
    df_out = df_in.copy()
    for c in cols:
        col = df_out[c].dropna()
        if col.empty:
            continue
        q1,q3 = col.quantile(0.25), col.quantile(0.75)
        iqr = q3-q1
        lower, upper = q1-1.5*iqr, q3+1.5*iqr
        df_out = df_out[(df_out[c] >= lower) & (df_out[c] <= upper)]
    return df_out
```

```python
cleaned_no_outliers = remove_iqr_outliers(cleaned, num_cols)
```

```python
cleaned.to_csv("/content/drive/MyDrive/NBPDCL/IAS cnsr LS  data (1).csv", index=False)
cleaned_no_outliers.to_csv("/content/cleaned_ias_no_outliers.csv", index=False)
```

```python
print("Before removing duplicates:", df.shape)
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)
```

```python
print(df.isnull().sum())
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
df['entry_ts'] = pd.to_datetime(df['entry_ts'], errors='coerce')
```

```python
col = 'wh_imp' # Changed from 'load' to an existing numerical column, please adjust as needed

Q1 = df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

df = df[(df[col] >= lower_limit) & (df[col] <= upper_limit)]
```

```python
df = df[df['wh_imp'] >= 0]
```

```python
df = df.reset_index(drop=True)
```

```python
print("Cleaning completed. Final shape:", df.shape)
```

```python
# ---------------------------
# 1. TIME-BASED FEATURES
# ---------------------------

# Drop columns that are entirely NaN, as these would cause df.dropna() to remove all rows
df = df.dropna(axis=1, how='all')

# Drop rows where 'entry_ts' is NaT (Not a Time) to ensure valid timestamps for feature creation
df = df.dropna(subset=['entry_ts'])

df['hour'] = df['entry_ts'].dt.hour
df['day'] = df['entry_ts'].dt.day
df['month'] = df['entry_ts'].dt.month
df['weekday'] = df['entry_ts'].dt.weekday
df['is_weekend'] = df['weekday'].apply(lambda x: 1 if x >= 5 else 0)

# ---------------------------
# 2. LAG FEATURES (PAST VALUES)
# ---------------------------

# Load at previous hour
df['load_lag_1'] = df['wh_imp'].shift(1)

# Load at previous day (24 hours earlier)
df['load_lag_24'] = df['wh_imp'].shift(24)

# ---------------------------
# 3. MOVING AVERAGE FEATURES
# ---------------------------

df['rolling_mean_3'] = df['wh_imp'].rolling(window=3).mean()
df['rolling_mean_24'] = df['wh_imp'].rolling(window=24).mean()

# ---------------------------
# 4. DROP NA CREATED BY SHIFT
# ---------------------------
df = df.dropna().reset_index(drop=True)

print("Feature engineering completed. New shape:", df.shape)
```
